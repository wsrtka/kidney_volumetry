```python
import json
import pandas as pd
import numpy as np
```

```python
data = []
with open('results.jsonl') as f:
    for l in f:
        res = json.loads(l)
        data.append(res)
```

```python
df = pd.DataFrame(data)
df = df.astype({'example_id': np.int})
```

```python
df
```

```python
ref_df = pd.read_excel('../data/do projektow.xlsx')
ref_df = ref_df.iloc[1:]
ref_df = ref_df.T.reset_index(drop=True).T
ref_df.rename(columns={x: str(x) for x in ref_df.columns}, inplace=True)
ref_df = ref_df[~pd.isna(ref_df['0'])]
ref_df = ref_df.astype({'0': np.int})
ref_df
results_df = df.merge(ref_df, how='left', left_on='example_id', right_on='0')
```

```python
results_df
```

```python
results_df.to_excel('experiment_results.xlsx')
```
